```python
import yfinance as yf
import pandas as pd

tickers = ["^GSPC", "^IXIC", "^NSEI"]
start_date = "2020-01-01"
end_date = "2026-01-12"

stocks_df = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")
stocks_df.head()
```

```python
stocks_df_cleaned = stocks_df.dropna()
```

```python
print(stocks_df_cleaned)
```

```python
#stacking method
df_stacked = stocks_df_cleaned.stack(level=0, future_stack=True)

# Flatten the index so 'Date' and 'Ticker' become regular columns
df_final = df_stacked.reset_index()
df_final['Date'] = pd.to_datetime(df_final['Date']).dt.date

print(df_final.head())
```

```python
# Rename 'level_1' to 'Ticker' for clarity
df_final = df_final.rename(columns={'level_1': 'Ticker'})

print(df_final.columns)
```

```python
df_final
```

```python
df_final.rename_axis(None, axis=1, inplace=True)
```

```python
# Removes the ^ symbol from every entry in the Ticker column
df_final['Ticker'] = df_final['Ticker'].str.replace('^', '', regex=False)

# Check the result
print(df_final['Ticker'].unique())
```

```python
mycursor.execute("""
    CREATE TABLE stock_prices(
        `date` DATE ,
        Ticker VARCHAR(20),
        Open DECIMAL(18, 6),
        High DECIMAL(18, 6),
        Low DECIMAL(18, 6),
        Close DECIMAL(18, 6),
        Volume BIGINT
        
        )
""")
```

```python
data_to_insert = [tuple(x) for x in df_final[['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Volume']].values]

stock_query = """
INSERT INTO stock_prices (Date, Ticker, Open, High, Low, Close, Volume) 
VALUES (%s, %s, %s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE 
    Open = VALUES(Open), 
    High = VALUES(High), 
    Low = VALUES(Low), 
    Close = VALUES(Close), 
    Volume = VALUES(Volume)
"""

# 3. Execution
mycursor.executemany(stock_query, data_to_insert)
mydb.commit()
print("Success! Data inserted.")
```

```python
query_1= "select * from stock_prices"
mycursor.execute(query_1)
all_data= mycursor.fetchall()
print("all stock prices")
for row in all_data:
    print(f" `date`: {row[0]} | Ticker: {row[1]} | Open:{row[2]} | High:{row[3]} | Low:{row[4]} |Close:{row[5]}")
```

```python
#Query 2 - highest closing price for NASDAQ (^IXIC)
```

```python
query_2 = """
SELECT 
    Date, 
    Ticker, 
    Close
FROM stock_prices 
WHERE Ticker = 'IXIC' 
ORDER BY Close DESC 
LIMIT 1
"""

mycursor.execute(query_2)
all_data = mycursor.fetchall()

print("--- Highest Close Value ---")
for row in all_data:
    print(f"Date: {row[0]} | Ticker: {row[1]} | Close: {row[2]}")
```

```python
#Query 3 -  top 5 days with highest price difference   (high - low) for S&P 500 (^GSPC)
```

```python
query_3 = """
SELECT 
    Date, 
    Ticker, 
    (High-Low) AS Difference 
FROM stock_prices 
WHERE Ticker = 'GSPC' 
ORDER BY Difference DESC 
LIMIT 1
"""

mycursor.execute(query_3)
High_Differences = mycursor.fetchall()

print("--- High_Differences ---")
for row in High_Differences:
    print(f"Date: {row[0]} | Ticker: {row[1]} | Difference: {row[2]}")
```

```python
#Query 4 -Get monthly average closing price for each ticker
```

```python
query_4 = """
SELECT 
    DATE_FORMAT(Date, '%Y-%m') AS Month, 
    Ticker, 
    AVG(Close) AS Avg_close
FROM stock_prices 
GROUP BY Month, Ticker
ORDER BY Month DESC
"""
mycursor.execute(query_4)
Avg_close = mycursor.fetchall()

print("--- Avg_close per month ---")
for row in Avg_close:
    print(f"Date: {row[0]} | Ticker: {row[1]} | Avg_close: {row[2]}")
```

```python
query_5 = """
SELECT 
    Ticker, 
    AVG(Volume) AS Avg_Volume
FROM stock_prices 
WHERE Ticker = 'NSEI'
GROUP BY Ticker
"""
mycursor.execute(query_5)
Avg_Volume = mycursor.fetchall()

print("--- Avg_Volume ---")
for row in Avg_Volume:
    print(f" Ticker: {row[0]} | Avg_Volume: {row[1]}")
```
